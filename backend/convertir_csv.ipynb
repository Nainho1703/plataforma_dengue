```python
import pandas as pd

# 1. Configuración del archivo
archivo_excel = 'data/DHF 2025_01Jul_29Nov by BMA.xls' # <--- Tu archivo Excel

# IMPORTANTE: Necesitas saber el nombre de las pestañas en tu Excel.
# Ábrelo y mira abajo. Supongamos que se llaman "Data" y "Codes".
# Si no lo sabes, pandas puede leer por posición (0 para la primera, 1 para la segunda).

print("Leyendo archivo Excel...")

# Leer la hoja de DATOS (asumiendo que es la primera pestaña, sheet_name=0)
# Mantenemos header=4 porque la estructura visual es la misma
df = pd.read_excel(archivo_excel, sheet_name=1, header=4) 
df
```

```python

# Leer la hoja de CÓDIGOS (asumiendo que es la segunda pestaña, sheet_name=1)
# O si sabes el nombre: sheet_name='code_subdistrict'
df_codes = pd.read_excel(archivo_excel, sheet_name=0)

# --- EL RESTO DEL SCRIPT ES IGUAL QUE ANTES ---


week_cols = df.columns[3:] # Las columnas de datos empiezan en la 4ta posición
week_nums = df.iloc[0, 3:].astype(str).str.replace('.0', '', regex=False)
mapa_nombres = {col: num for col, num in zip(week_cols, week_nums)}

df_clean = df.drop(0).rename(columns=mapa_nombres)
df_clean
```

```python
import geopandas as gpd
import folium
# Crear ID compatible con GeoJSON (TH + 6 dígitos)
# Ejemplo: 100101.0 -> "TH100101"

RUTA_GEOJSON = 'data/tha_admbnda_adm3_rtsd_20220121.geojson'
df_clean['Sub-district code'] = pd.to_numeric(df_clean['Sub-district code'], errors='coerce')
df_clean = df_clean.dropna(subset=['Sub-district code'])
df_clean['ID_MAPA'] = 'TH' + df_clean['Sub-district code'].astype(int).astype(str)
SEMANA_ELEGIDA = '30'
# Seleccionar solo la semana elegida y el ID
if SEMANA_ELEGIDA not in df_clean.columns:
    print(f"¡Error! La semana {SEMANA_ELEGIDA} no existe en los datos. Semanas disponibles: {week_nums.unique()}")
else:
    df_mapa = df_clean[['ID_MAPA', SEMANA_ELEGIDA]].copy()
    df_mapa.columns = ['ID_MAPA', 'Casos'] # Renombrar para facilitar
    df_mapa['Casos'] = pd.to_numeric(df_mapa['Casos'], errors='coerce').fillna(0)

    # --- 3. CARGA DEL MAPA (GEOJSON) ---
    print("Cargando GeoJSON...")
    try:
        gdf = gpd.read_file(RUTA_GEOJSON)
        
        # Filtrar solo Bangkok (Código provincia TH10)
        # En HDX la columna suele ser 'ADM1_PCODE'
        gdf_bangkok = gdf[gdf['ADM1_PCODE'] == 'TH10'].copy()
        
        # --- 4. UNIÓN DE DATOS ---
        # Unimos el mapa con tus datos de casos
        # GeoJSON usa 'ADM3_PCODE' para el código de sub-distrito
        gdf_final = gdf_bangkok.merge(
            df_mapa,
            left_on='ADM3_PCODE',
            right_on='ID_MAPA',
            how='left'
        )
        gdf_final['Casos'] = gdf_final['Casos'].fillna(0)

        # --- 5. VISUALIZACIÓN ---
        print(f"Generando mapa para la Semana {SEMANA_ELEGIDA}...")
        
        # Crear mapa centrado en Bangkok
        m = folium.Map(location=[13.7563, 100.5018], zoom_start=11)

        # Capa de Coropletas (Colores)
        folium.Choropleth(
            geo_data=gdf_final,
            data=gdf_final,
            columns=['ID_MAPA', 'Casos'],
            key_on='feature.properties.ADM3_PCODE', # Ojo: propiedad del GeoJSON original
            fill_color='YlOrRd', # Amarillo a Rojo
            fill_opacity=0.7,
            line_opacity=0.2,
            legend_name=f'Casos de Dengue - Semana {SEMANA_ELEGIDA}',
            highlight=True
        ).add_to(m)

        # Tooltips (para ver info al pasar el mouse)
        folium.GeoJson(
            gdf_final,
            style_function=lambda x: {'fillColor': '#00000000', 'color': 'black', 'weight': 0.1},
            tooltip=folium.GeoJsonTooltip(
                fields=['ADM3_EN', 'Casos'],
                aliases=['Sub-distrito:', 'Casos:'],
                localize=True
            )
        ).add_to(m)

        # MOSTRAR MAPA EN EL NOTEBOOK
        display(m)

    except Exception as e:
        print(f"Error: {e}")
        print("Asegúrate de que la ruta al archivo GeoJSON es correcta y es el archivo de nivel 3 (adm3).")

```
